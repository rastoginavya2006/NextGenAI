```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("headbrain.csv")
```

```python
df
```

```python
X = df['Head Size(cm^3)'].values
y = df['Brain Weight(grams)'].values
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y)
plt.show()
```

```python
mean_x = np.mean(X)
mean_y = np.mean(y)
```

```python
numer = 0
denom = 0
n = len(X)

for i in range(n):
    numer += (y[i] - mean_y) * (X[i] - mean_x)
    denom += (X[i] - mean_x) ** 2

slope = numer / denom
intercept = mean_y - (slope * mean_x)
```

```python
print("Slope:",slope)
print("Intercept:",intercept)
```

```python
predictions = []
for i in range(n):
    y_pred = slope * X[i] + intercept
    predictions.append(int(y_pred))
```

```python
# Predicted Values
print(predictions[:5])
```

```python
# Actual Values
print(y[:5])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y)
plt.plot(X, predictions, 'r')
plt.show()
```

```python
# MSE - Mean Squared Error
error = 0
for i in range(n):
    error += (predictions[i] - y[i]) ** 2

mse = error/n
print("MSE:",mse)
```

```python
# RMSE
np.sqrt(mse)
```

# **Gradient Descent Based Slope and Intercept**

```python
# Normalization
from sklearn.preprocessing import MinMaxScaler
```

```python
X.shape
```

```python
# Making our data 2D, so that we can pass in MinMaxScaler
X = np.reshape(X, (-1,1))
y = np.reshape(y, (-1,1))
```

```python
X.shape
```

```python
minmax = MinMaxScaler()
scale_x = minmax.fit_transform(X)
scale_y = minmax.fit_transform(y)
```

```python
print(scale_x[:5])
```

```python
def gradientDescent(epochs,alpha):
    slope, inter = 0,0
    for i in range(epochs):
        y_pred = scale_x * slope + inter # y = mx + c
        loss = y_pred - scale_y
        gradSlope = (2/n) * np.dot(loss.T, scale_x)
        gradInter = (2/n) * np.sum(loss)
        slope = slope - gradSlope * alpha
        inter = inter - gradInter * alpha
    return slope, inter    
```

```python
epochs = 7000
alpha = 0.01
slope, inter = gradientDescent(epochs, alpha)
```

```python
print("New Slope:",slope)
print("New Inter:",inter)
```

```python
predictions = []
for i in range(n):
    y_pred = slope[0][0] * scale_x[i] + inter
    predictions.append(y_pred)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(scale_x, scale_y)
plt.plot(scale_x, predictions, 'r')
plt.show()
```
